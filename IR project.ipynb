```python
import nltk
from nltk import word_tokenize
from nltk.corpus import stopwords
import collections
import pandas as pd
import math
from numpy import dot
from numpy.linalg import norm
import os
```

```python
path = "D:\\Programming\\4th year\\1st Term\\IR (Information Storage and Retrieval)\\IR project\DocumentCollection"
docs = {}
for i in range(1,11):
  f = open(path + '\\' + str(i) + ".txt" , "r")
  docs[i] = f.read().lower()
docs
```

```python
toknization = {}
for i in docs:
  toknization[i] = word_tokenize(docs[i].lower())
toknization
```

```python
stop_words = set(stopwords.words('english'))
stop_words.remove("in")
stop_words.remove("to")
stop_words.remove("where")

filtered_Sent = collections.defaultdict(lambda : [])

for i in toknization:
  for w in toknization[i]: 
    if w not in stop_words:
      filtered_Sent[i].append(w)

for i in filtered_Sent:
  print(i,filtered_Sent[i])
```

```python
words = []
for i in filtered_Sent:
  for w in filtered_Sent[i]:
    words.append(w)


uniqe_words = list(set(words))
uniqe_words.sort()
print(len(uniqe_words))
print(uniqe_words)
```

```python
pos_index = collections.defaultdict(lambda : [0,{}])
for w in uniqe_words:
  for doc in toknization:
    if w in toknization[doc]:
      pos_index[w][0] += 1
      pos_index[w][1][doc] = []

        
print(pos_index)        

for un_W in pos_index:
  for doc in toknization: #doc = [asdf,asdfas,dfasdf,asdf,sadf]
    if un_W in toknization[doc]:
      counter = 0
      while (counter < len(toknization[doc])):
        if un_W == toknization[doc][counter]:
          pos_index[un_W][1][doc].append(counter+1)
        counter+=1

#for w in pos_index:
#  print(w,pos_index[w])
#print("***********************************")
#for w in toknization:
#  print(w,toknization[w])
pos_index
```

```python
w = "antony"
d = 1
def Compute_TF(word,docID):
    if docID not in list(pos_index[word][1]):
      return 0
    else:
      return len(pos_index[word][1][docID])
```

```python
def Compute_WeTF(word,docID):
    wtfreq = Compute_TF(word,docID)
    if (wtfreq > 0):
        return (1 + math.log(wtfreq,10))
    elif (wtfreq == 0):
        return 0
```

```python
IDF_matrix = {}
for word in pos_index:
  row = []
  row.append(pos_index[word][0])
  row.append(Compute_IDF(word))
  IDF_matrix[word] = row


IDF_matrix_table = pd.DataFrame(IDF_matrix.values(),columns=["Doc_freq" , "IDF"] , index=uniqe_words)
IDF_matrix_table
```

```python
def Compute_TF_IDF(word,docID):
    return Compute_IDF(word)*Compute_WeTF(word,docID)
```

```python
TF_IDF_matrix = []
for word in pos_index:
    TF_IDF_each_term = []
    for docID in docs:     
        TF_IDF_each_term.append(Compute_TF_IDF(word,docID))
    TF_IDF_matrix.append(TF_IDF_each_term)

TF_IDF_matrix_table = pd.DataFrame(TF_IDF_matrix,columns=docs , index=uniqe_words)
TF_IDF_matrix_table
```

```python
length = {}
for docID in filtered_Sent:
    TF_IDF_sum = 0
    for word in filtered_Sent[docID]:
        TF_IDF_sum += math.pow(Compute_TF_IDF(word,docID),2)
    length[docID] = math.sqrt(TF_IDF_sum)

    
length_table = pd.DataFrame(length.values(),columns=["Length"] , index=filtered_Sent)
length_table
```

```python
NORM_TF_IDF_matrix = []
for word in pos_index:
    TF_IDF_each_term = []
    for docID in docs:     
        TF_IDF_each_term.append(Compute_TF_IDF(word,docID)/length[docID])
    NORM_TF_IDF_matrix.append(TF_IDF_each_term)

NORM_TF_IDF_matrix_table = pd.DataFrame(NORM_TF_IDF_matrix,columns=docs, index=uniqe_words)
NORM_TF_IDF_matrix_table
```

```python
sent = "antony brutus"

def get_mathched_docs(sent):
    tokenized_sent = word_tokenize(sent.lower()) 
    # check if all words in our docs or not
    notfound = False
    for item in tokenized_sent:
        if item not in pos_index:
            notfound = True
            break
    if(notfound):
        return 0
    #check if all words in same doc or not
    commen_docs = set(list(pos_index[tokenized_sent[0]][1]))
    for item in tokenized_sent:
        item_set = set(list(pos_index[item][1]))
        commen_docs = commen_docs.intersection(item_set)
    if(len(commen_docs) == 0):
        return 0
    commen_docs = list(commen_docs)
       
    matched_docs = []
    
    orders = {}
    for doc in commen_docs:
        postions = []
        for word in tokenized_sent:
            for postion in pos_index[word][1][doc]:
                postions.append(postion)
        orders[doc] = postions  
    
    
    for id in orders:
        if (orders[id] == list(range(min(orders[id]),max(orders[id])+1))):
            matched_docs.append(id)
    
    if(matched_docs == 0):
        return 0
    
    return matched_docs
    
    
get_mathched_docs(sent)
```

```python
sent = input("Enter Query: ")
matched_docs = get_mathched_docs(sent)
if(matched_docs):
    toknized_sent = word_tokenize(sent.lower()) 

    filtered_sent = []

    for word in toknized_sent:
        if(word not in stop_words):
            filtered_sent.append(word)

    uni_words = list(set(filtered_sent))
    uni_words.sort()
    content = {}
    vector = []
    summation = 0
    for word in uni_words:
        row = []
        #TF
        if(word in uni_words):
            row.append(toknized_sent.count(word))
        else:
            row.append(0)
        #WeTF
        if(row[0] > 0):
            row.append(1+math.log(row[0],10))
        else:
            row.append(0)
        #IDF
        row.append(IDF_matrix[word][1])
        #TF_IDF
        row.append(row[1]*row[2])
        #sum_length
        summation += math.pow(row[3],2)
        content[word] = row

    #length
    length = math.sqrt(summation)
    for word in content:
        norm = content[word][3]/length
        content[word].append(norm)
        vector.append(norm)

    content_table = pd.DataFrame(content.values(),columns=["TF","Weighted_TF","IDF","TF_IDF","Norm_TF_IDF"] , index=content.keys())
    display(content_table)

    print("query length: ",length,"\n")

    #cosine_simlarity
    
    cosine_sim = {}
    for doc_id in matched_docs:
        doc_vector = []
        for word in content:
            doc_vector.append(vector_form_for_docs[doc_id][uniqe_words.index(word)])
        cosine_sim[doc_id] = Cosine_simalrty(vector,doc_vector)
        
    
    Ranked_docs = sorted(cosine_sim.items(),key=lambda x:x[1] , reverse = True)
    Ranked_docs = dict(Ranked_docs)
    
    for doc in cosine_sim:
        print("cosine simlarity ( q ,",str(doc)+".txt) :",cosine_sim[doc])
        
    print("\n matched docs: ", list(Ranked_docs.keys()))
else:
    print("Not Found")
```
